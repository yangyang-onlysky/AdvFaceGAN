```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 数据
x = [2, 3, 4, 5, 6, 7, 8]
y1 = [0.978145, 0.963782, 0.944190, 0.918466, 0.900313, 0.874187, 0.845629]  # ssim 数据
y2 = [0.682487, 0.859459, 0.936999, 0.970588, 0.983289, 0.988636, 0.991811]  # MobileFace ASR 数据
y3 = [0.741979, 0.879512, 0.933824, 0.952039, 0.967079, 0.967580, 0.972426]  # SphereFace ASR 数据
y4 = [0.688168, 0.854947, 0.922627, 0.952707, 0.966076, 0.970922, 0.974098]  # CosFace ASR 数据
y5 = [0.439171, 0.639873, 0.772894, 0.875334, 0.897894, 0.934492, 0.948362]  # Facenet ASR 数据
# 创建一个图形对象和一个子图对象
fig, ax1 = plt.subplots()

# 绘制左侧y轴的数据
ax1.plot(x, y1, 'b-', marker='o', linewidth=4)
ax1.set_xlabel('UpperLimit Of Perturbation $ε$', fontsize=16)
ax1.set_ylabel('SSIM', color='b', fontsize=16)
ax1.tick_params('y', colors='b')

# 创建右侧y轴并绘制数据
ax2 = ax1.twinx()
ax2.plot(x, y2, 'r-', marker=None, linewidth=4)
ax2.plot(x, y3, 'm-', marker=None, linewidth=4)
ax2.plot(x, y4, 'y-', marker=None, linewidth=4)
ax2.plot(x, y5, 'k-', marker=None, linewidth=4)
ax2.set_ylabel('${ASR}_{target}$', color='r', fontsize=16)
ax2.tick_params('y', colors='r')

# 添加图例
fig.tight_layout()
fig.legend(("SSIM", "MobileFace ${ASR}_{target}$","SphereFace ${ASR}_{target}$","CosFace ${ASR}_{target}$","Facenet ${ASR}_{target}$",),loc='upper right')

# 显示图形
plt.title('')
plt.show()
```
